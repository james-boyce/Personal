```python
import pandas as pd

# names of files to read from
r_filenameTSV1 = '../Data/amazon_reviews_us_Grocery_v1_00.tsv'

# names of files to read from
r_filenameTSV2 = '../Data/amazon_reviews_us_Pet_Products_v1_00.tsv'

# names of files to read from
r_filenameTSV3 = '../Data/amazon_reviews_us_Sports_v1_00.tsv'

# names of files to read from
r_filenameTSV4 = '../Data/amazon_reviews_us_Beauty_v1_00.tsv'

# names of files to read from
r_filenameTSV5 = '../Data/amazon_reviews_us_Home_v1_00.tsv'


# read the data
grocery = pd.read_csv(r_filenameTSV1, sep='\t', error_bad_lines=False)
pet = pd.read_csv(r_filenameTSV2, sep='\t', error_bad_lines=False)
sport = pd.read_csv(r_filenameTSV3, sep='\t', error_bad_lines=False)
beauty = pd.read_csv(r_filenameTSV4, sep='\t', error_bad_lines=False)
home = pd.read_csv(r_filenameTSV5, sep='\t', error_bad_lines=False)
```

```python
frames = [grocery, pet, sport, beauty, home]
```

```python
result = pd.concat(frames)
```

```python
df = result.reset_index(drop=True)
```

```python
## Obtain customer activity from entire dataset
activity = df.groupby('customer_id')
users = activity['review_id'].agg('count')
user_df = pd.DataFrame({'customer_id':users.index, 'Activity':users.values})
user_df
```

```python
#Drop duplicates in dataset and reset index
df.drop_duplicates(subset = ['product_id', 'star_rating', 'review_headline', 'review_body'], keep='first', inplace=True)
df = df.reset_index()
```

```python
#Remove Null Entries
clean_df = df.dropna()
```

```python
#Append list to convert string data to datetime format
import datetime
new_date = []
for row in clean_df['review_date']:
    temp = datetime.datetime.strptime(row, "%Y-%m-%d")
    new_date.append(temp)
```

```python
#Add to datframe
date_df = pd.DataFrame(new_date)
```

```python
#Remove old date
new_df = clean_df.drop(["review_date"], axis = 1)
```

```python
#Add new date 
new_df.insert(14, 'new_date', new_date)
new_df
```

```python
##Find earliest review date on each product
#Group products together
product_dates = new_df.groupby('product_id')
```

```python
#Create list of the earliest review date of each product 
min_date = product_dates['new_date'].agg('min')
```

```python
#Convert to dataframe
min_df = min_date.to_frame(name='Min').reset_index()
```

```python
#Add year to the min_df
min_df['Year'] = pd.DatetimeIndex(min_df['Min']).year
```

```python
#Filter for only products with the earliest review in 2015 (recent products)
#Reset the index
recent_products = min_df[min_df['Year'] == 2015]
recent_products = recent_products.reset_index(drop=True)
```

```python
#Create a reduced dataframe with only the products in recent_product dataframe
reduced_df = pd.merge(recent_products, new_df, on="product_id", how='left')
reduced_df
```

```python
##Find Number of reviews per product
#Group by products
product_reviews = reduced_df.groupby('product_id')
```

```python
#Count the number of review id's (reviews) per product
product_review_count = product_reviews['review_id'].agg('count')
product_review_count
```

```python
#Convert to dataframe and reset index
count_df = product_review_count.to_frame().reset_index()
count_df
```

```python
#Filter out entries with less than 10 reviews
filter_count = count_df[count_df['review_id'] >= 10]
filter_count
```

```python
#Create a new reduced dataframe with products that have 10 or more reviews
filter_df = pd.merge(filter_count, reduced_df, on = 'product_id', how = 'left')
filter_df
```

```python
#Drop columns with no use
filter_df = filter_df.drop(["index", "review_id_x", "Min", "Year", "marketplace"], axis = 1)
```

```python
#Function to find the helpfulness index
def helpIndex(row):
    if row['total_votes'] == 0:
        val = 0
    else:
        val = (row['helpful_votes'] / row['total_votes']) 
    return val
```

```python
#Add index to dataframe
filter_df['helpfulness_index'] = filter_df.apply(helpIndex, axis=1)
```

```python
##Find Average index for each customer
#Group by customer
customer_index = filter_df.groupby('customer_id')
```

```python
#Average the index per customer
index_avg = customer_index['helpfulness_index'].agg('mean')
index_avg
```

```python
#Convert averages to a dataframe and reset index
average_df = index_avg.to_frame(name='index_avg').reset_index()
average_df
```

```python
#Create new dataframe with the averages added for each user
veri_index_df = pd.merge(average_df, filter_df, on = 'customer_id', how = 'left')
veri_index_df
```

```python
#Add Activity to remaining users
reviewer_df = pd.merge(user_df, veri_index_df, on="customer_id", how='right')
reviewer_df
```

```python
#Function to calculate user reputation
def reputation(row):
    val = (row['index_avg'] * row['Activity']) 
    return val
```

```python
#Add reputation to dataframe
reviewer_df['reputation'] = reviewer_df.apply(reputation, axis=1)
```

```python
#Add Review Length
reviewer_df['Text_Length'] = reviewer_df['review_body'].str.split().str.len()
reviewer_df
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('Research_Dataset.xlsx', engine='xlsxwriter')

# Don't convert url-like strings to urls.
writer.book.strings_to_urls = False

# Convert the dataframe to an XlsxWriter Excel object.
reviewer_df.to_excel(writer, sheet_name='Reviews', index = False)

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
